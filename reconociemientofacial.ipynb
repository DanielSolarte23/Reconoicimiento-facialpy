```python
import cv2
import tensorflow as tf
import numpy as np

# Carga el modelo preentrenado para detección de rostros
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Carga el modelo de emociones entrenado
model = tf.keras.models.load_model('modelo_emociones.h5')

# Etiquetas de las emociones (para FER-2013)
emotion_labels = ['enojado', 'disgustado', 'Temor', 'Feliz', 'Triste', 'Sorpresa', 'Neutral']

# Función para predecir la emoción usando el modelo entrenado
def predict_emotion(face):
    # Preprocesa la imagen del rostro
    face_resized = cv2.resize(face, (48, 48))  # Ajusta el tamaño según lo que espera tu modelo
    face_normalized = face_resized / 255.0  # Normaliza los valores a [0, 1]
    face_expanded = np.expand_dims(face_normalized, axis=0)  # Agrega una dimensión para el batch
    face_expanded = np.expand_dims(face_expanded, axis=-1)  # Agrega la dimensión del canal (1 para escala de grises)
    
    # Predice la emoción
    predictions = model.predict(face_expanded)
    emotion_index = np.argmax(predictions)  # Índice de la emoción más probable
    return emotion_labels[emotion_index]

# Inicia la cámara
cap = cv2.VideoCapture(1)

if not cap.isOpened():
    print("Error: No se puede acceder a la cámara.")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: No se pudo capturar el fotograma.")
        break

    # Convierte el frame a escala de grises (necesario para la detección de rostros)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detecta rostros en el fotograma
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        # Dibuja un rectángulo alrededor del rostro detectado
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

        # Extrae el rostro para predecir la emoción
        face_roi = gray[y:y+h, x:x+w]

        # Predice la emoción
        emotion = predict_emotion(face_roi)

        # Muestra la emoción sobre el rostro detectado
        cv2.putText(
            frame, emotion, 
            (x, y - 10),  # Coordenadas del texto (encima del rostro)
            cv2.FONT_HERSHEY_SIMPLEX, 
            0.6,  # Tamaño del texto
            (255, 255, 255),  # Color del texto (blanco)
            2  # Grosor del texto
        )

    # Muestra el fotograma en una ventana
    cv2.imshow("Deteccion de Rostros", frame)

    # Detiene el bucle al presionar 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Libera los recursos
cap.release()
cv2.destroyAllWindows()
```
